### 2. Use crosstabs to show the number of moovies per country

```python
data.reset_index(inplace=True)
```

```python
pd.crosstab(data.country,data.color,values=data.index, aggfunc='count',margins=True,margins_name="Total")
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab([data.director_name,data.genres,],data.country,rownames=['Director','Genres'], colnames=["Country"],
            dropna=True,margins=True)
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(data.duration,data.color,values=data.duration, aggfunc='mean',margins=True,margins_name="Total",dropna=True)
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
pd.crosstab(data.director_name,data.director_facebook_likes,values=data.actor_2_facebook_likes, aggfunc='mean',margins=True,margins_name="Total",dropna=True)
```
