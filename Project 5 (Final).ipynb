```python
#The stake holders want to know, for each Department, the number of employees and how much money is spent on salaries.

conditions = [
    (IBM['Department'] == 'Human Resources'),
    (IBM['Department'] == 'Research & Development'),
    (IBM['Department'] == 'Sales')
    ]

# create a list of the values we want to assign for each condition
values = [(IBM['Department'] == 'Human Resources').sum(), (IBM['Department'] == 'Research & Development').sum(), (IBM['Department'] == 'Sales').sum()] 

# create a new column and use np.select to assign values to it using our lists as arguments

IBM['Number_of_employees'] = np.select(conditions, values) 
```

```python
#For each Department, the monthly Income per department whatever the Attrition is:
x = IBM.pivot_table(index=["Department", "Number_of_employees"], values=["MonthlyIncome"], aggfunc=sum)
plt.title('Total Salaries by Department')
plt.ylabel("Salaries per Month")
x.plot(kind='bar')
```
