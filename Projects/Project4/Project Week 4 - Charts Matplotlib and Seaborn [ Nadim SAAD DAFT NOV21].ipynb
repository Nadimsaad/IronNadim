```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
diamonds = pd.read_csv(r'/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/diamonds_clean.csv')
diamonds = diamonds.drop('Unnamed: 0', axis=1)
```

```python
diamonds.groupby(['color'])['carat', 'clarity_scale', 'table','price', 'depth'].mean()
```

```python
# LINE CHART (MATPLOTLIB) MEAN PRICE BY COLOR

# Find the mean price for each color
diamond_mean_price_color = diamonds['price'].groupby(diamonds['color']).mean()

# Plot the mean price per color in the current axes
diamond_mean_price_color.plot(kind = 'line',linewidth=2)

# Add labels, title, etc
plt.ylabel('Mean price per color')
plt.xlabel('Colors')
plt.title('Mean price per diamond color')
plt.xticks(range(7), ['J','I','H','G','F','E','D'])

# Save the figure
plt.savefig('/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/Figures/mean_prices_per_diamond_color.png',bbox_inches='tight')
```

```python
# BAR CHART (MATPLOTLIB) COUNT PER DIAMOND COLOR

# Find the diamonds count per color
diamond_count_per_color = diamonds['color'].value_counts()

# Plot the diamonds count per color axes
diamond_count_per_color.plot(kind = 'bar', alpha = 0.5)

# Add labels, title, etc
plt.ylabel('Diamonds count')
plt.xlabel('Diamonds colors')
plt.title('Diamonds count per color')
plt.xticks(range(7), ['J','I','H','G','F','E','D'])

# Save the figure
plt.savefig('/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/Figures/Diamonds_count_per_color.png', bbox_inches='tight')
```

```python
# PIE CHART (MATPLOTLIB) COUNT PER DIAMOND CLARITY

# Find the diamonds count per clarity
diamond_count_per_clarity = diamonds['clarity'].value_counts()

# Plot the diamonds count per clarity type pie chart
mylabels = ['1','2','3','4','5','6','7']
myexplode = [0, 0, 0, 0, 0, 0, 0.4]
plt.title('Diamonds count per clarity')
plt.pie(diamond_count_per_color, labels = mylabels , explode = myexplode, autopct='%1.1f%%')
plt.legend(title = "Diamonds clarity:")
plt.show() 

# Save the figure
#plt.savefig('/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/Figures/Diamonds_count_per_clarity_(Pie_Chart).png', bbox_inches='tight')
plt.savefig('/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/Figures/Diamonds_count_per_clarity_(Pie_Chart).png')
```

```python
# BOXPLOTCHART (MATPLOTLIB)

plt.boxplot(diamonds['price'])
plt.title('Price Boxplot')
plt.ylabel('Price')

plt.savefig('/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/Figures/Diamonds_Price_BoxPlot.png', bbox_inches='tight')
```

```python
# SCATTER CHART (MATPLOTLIB)

plt.figure(figsize=(10,8))
plt.scatter(x=diamonds['carat'],y=diamonds['price'], color='blue')
plt.title('Diamond Carat Type vs. Price')
plt.xlabel('Carat Type')
plt.ylabel('Diamond Price')

# Save the figure
plt.savefig('/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/Figures/Diamonds_Carat_vs_Price.png', bbox_inches='tight')
```

```python
# BOXPLOT

diamonds.boxplot()
```

```python
# COLORED HORIZONTAL BAR CHART (MATPLOTLIB)
colorgrp = diamonds.groupby('cut')['color_scale'].value_counts().unstack()
colorgrp.plot.barh(stacked=True)

plt.ylabel('Diamonds count')
plt.xlabel('Diamonds colors')
plt.title('Diamonds Colors Scale per Cut Type')

# Save the figure
plt.savefig('/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/Figures/Diamonds_Color_Scale_per_Cut_Type.png', bbox_inches='tight')
```

```python
# PIE CHART (SEABORN) 
 
#define data
x = diamonds.groupby(['color','color_scale']).agg('count').reset_index()

#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]

#create pie chart
plt.pie(x['color_scale'], labels=x['color'], autopct='%.0f%%')
plt.title('Share per Diamond Color Type')
plt.show()

# Save the figure
plt.savefig('/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/Figures/Share per Diamond Color Type.png', bbox_inches='tight')
```

```python
# SCATTERED PLOT (SEABORN) BAR CHART

sns.scatterplot(data=diamonds, x='carat',y='price', hue='color').set_title('Diamond Price compared to the Carat')

# Save the figure
plt.savefig('/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/Figures/Diamonds_Price_Compared_to_Carat.png', bbox_inches='tight')
```

```python
# HISTOGRAM PLOT (SEABORN) 

sns.distplot(diamonds['clarity_scale'], bins = 20, kde = False, color = 'red')

plt.ylabel('Diamonds count')
plt.xlabel('Diamonds clarity Scale')
plt.title('Diamonds Count per Clarity Type')

# Show the plot
plt.show()

# Save the figure
plt.savefig('/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/Figures/Diamonds_Count_Per_Clarity_Type.png', bbox_inches='tight')
```

```python
# VIOLIN PLOT (SEABORN)

# Set figure size
plt.figure(figsize = (15,8))

# Violin plot
sns.violinplot("price", data = diamonds)

plt.title('Diamonds Count Per Price ')

# Show the plot
plt.show()

# Save the figure
plt.savefig('/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/Figures/Diamonds_Count_Per_Price.png', bbox_inches='tight')
```

```python
# JOIN PLOT (SEABORN)

# Joinplot
sns.jointplot(x = 'carat', y = 'price', data = diamonds)
plt.title('Relationship Between Carat and Price ')

# Show the plot
plt.show()

# Save the figure
plt.savefig('/Users/nadimsaad/Desktop/IRONHACK/Module 2/Week 4/Day 20/Project 4/Figures/Relationship_Between_Price_and_Carat.png')
```
